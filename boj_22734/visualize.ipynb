```python
in_filepath = './sample4.in'

with open(in_filepath, 'r') as f:
    lines = iter(f.readlines())
    
    
def input():
    return next(lines)

N = int(input())
circles = []
for _ in range(N):
    r, x, y = map(float, input().split())
    circles.append((x, y, r))
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=[12, 12]) # note we must use plt.subplots, not plt.subplot

for i, (x, y, r) in enumerate(circles):
    c = plt.Circle((x, y), r)
    ax.add_patch(c)
    ax.text(x-0.5, y-0.5, str(i))

ax.set_aspect('equal', adjustable='datalim')
ax.plot()
plt.show()
```
